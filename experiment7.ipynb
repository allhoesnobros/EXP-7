```python
s = "madam"
i,j = 0, len(s)-1
is_palindrome = True
while i<j:
  if s[i]!=s[j]:
    is_palindrome = False
    break
  i+=1
  j-=1
if is_palindrome:
  print("palindrome")
else:
  print("not palindrome")
```

```python
#search an element in the list
nums = [10, 20, 30, 40, 50]
key = int(input("Enter the element to search: "))
i = 0
while i< len(nums):
  if nums[i]==key:
    print("Element found at index", i)
    break
  i += 1
else:
  print("Element not found")
```
